# Pandas Series

```python
#-->it is a one dimension array holding data of any type
#--> it is like a column in a table

import pandas as pd

a =[1,3,5,6,7,8]

myvar = pd.Series(a)
print(myvar)
```

```python
#Labels
#-->If nothing else is specified, the values are labeled with their index number. 
#   First value has index 0, second value has index 1 etc.

# label is used to access the value.....
#eg:

print(myvar)
```

```python
# access using index number

print(myvar[0])
print(myvar[4])
```

```python
#Create label--> with the 'index' argument,we can create  label

#eg::

import pandas as pd

a = [10,20,30]

myseries = pd.Series(a, index=["X","Y","Z"])
print(myseries)
```

```python
#access value using label
print(myseries["X"])
print(myseries["Z"])
```

```python
#  'Key/value' Object as a series.......

import pandas as pd

calories = {"day1":420,"day2":300,"day3":210}
print(pd.Series(calories))
```

```python
#only two value
print(pd.Series(calories,index=["day1","day2"]))
```

```python
#accessing values using key
print(calories["day3"])
```

# .............Cleaning Data..........

```python
# Data cleaning means fixing bad data in datasets
'''
Bad data could be::
1. Empty cells
2. Data in wrong format
3. Wrong data
4. Duplicates
'''
```

# 1. Remove/Replace Empty cell

```python
#Remove empty cell or null values
#--> one way to deal with empty cell is to remove rows which contain null value.this is ok with dataset 
#    containing large data.

import pandas as pd

df = pd.read_csv('data_for_cleaning.csv')
print(df.to_string())
```

```python
# to remove this empty cell we can use 'dropna()' method

new_df = df.dropna()
print(new_df)
```

```python
#the dropna() method doesn't change original dataframe.
#if we want to change original too, we can use dropna(inplace=True)

df.dropna(inplace=True)
print(df.to_string())
```

```python
#Replace empty cell
#--> fillna() method is used to insert value in the empty cell.

import pandas as pd

df = pd.read_csv('data_for_cleaning1.csv')
#print(df)
df.fillna(130,inplace=True)
print(df)
```

```python
#replace only for specified cloumn

import pandas as pd

df = pd.read_csv('data_for_cleaning.csv')

df["Date"].fillna(130,inplace=True)
print(df)
```

```python
df.describe()
```

```python
# Replace using Mean,Median and Mode
# mean--> average value of that column
# median --> middle value of that column after sorted in ascending order
# mode --> the value that appear most frequently

#comman way to replace value is find mean,median and mode 
# pandas has mean(),median(),mode() method to calculate it.

import pandas as pd

df = pd.read_csv('data_for_cleaning.csv')

x = df['Calories'].mean()
#x = df["Calories"].median()
#x = df["Calories"].mode()[0]


df['Calories'].fillna(x,inplace=True)
print(df)
```

# 2. Clearing wrong format

```python
# to fix this we have two option.
# 1. remove this row
# 2. convert that column in same format

#-- convert into correct format
# here date is not in correct format in row 25

import pandas as pd

df = pd.read_csv('data_for_cleaning.csv')
df
```

```python
# pandas has to_datetime() method to convert all the column in same format

import pandas as pd

df = pd.read_csv('data_for_cleaning.csv')
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
# remove row which doesnt have same format

df.dropna(subset=['Date','Calories'],inplace=True)
df
```

# 3. Cleaning Wrong Data

```python
# as we can see in row 7 value is 450.but its not correct value so we need to replace this

df.loc[7,'Duration'] =10
df
```

```python
# replace value for large data set.for that we create some boundries to replace values
# if the value is less than 50, replace it with 50
#for that

import pandas as pd

df = pd.read_csv('data_for_cleaning.csv')

for x in df.index:
    if df.loc[x,'Duration'] <50:
        df.loc[x,'Duration'] = 50
df

```

```python
# if we want to remove values that is wrong.
# in that case also we have some boundries.
# we will remove values on duration column which is greater that 100.

import pandas as pd

df = pd.read_csv('data_for_cleaning.csv')

for x in df.index:
    if df.loc[x,'Duration'] >100:
        df.drop(x,inplace=True)
df
```

# 4. Remove Duplicate Values

```python
# we can simply delete duplicate rows
# to find the duplicate rows use duplicated() method 
# to remove duplicate row --> use drop_duplicates() method

import pandas as pd

df = pd.read_csv('data_for_cleaning.csv')

print(df.duplicated())
```

```python
# it shows that row 12 has duplicate row
#we have to remove that

df.drop_duplicates(inplace=True)
df
```
